```python
import pandas as pd
import plotly.graph_objs as go

# Read the CSV file into a pandas DataFrame
df = pd.read_csv("macronutrients.csv")

```

```python
df.head()
```

```python
#Rename

df = df.rename(columns={'2-3': 'Age 2-3', '4-8': 'Age 4-8', '9-13': 'Age 9-13', '14-18': 'Age 14-18'})
```

```python
import pandas as pd
import matplotlib.pyplot as plt


colors = ['#1f77b4', '#ff7f0e', '#2ca025']

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Add pie charts for each age group
for i, ax in enumerate(axs.flat):
    age_group = df.columns[i+2]
    explode = [0.1] * len(df)  # Creating an explosion effect for elevation
    ax.pie(df[age_group], labels=df['Macronutrients'], colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)
    ax.set_title(age_group, fontsize=14, fontweight='bold')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
